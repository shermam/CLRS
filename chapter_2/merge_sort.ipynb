```python
def merge(a: list[int], p: int, q: int, r: int):
  n1 = q - p + 1
  n2 = r - q
  left = [float('inf')] * (n1+1)
  right = [float('inf')] * (n2+1)
  for i in range(n1):
    left[i]=a[p+i]
  for j in range(n2):
    right[j]=a[q+1+j]
  i=0
  j=0
  for k in range(p,r):
    if left[i] <= right[j]:
      a[k]=left[i]
      i+=1
    else:
      a[k]=right[j]
      j+=1
```

```python
a = [1,3,5,7,2,4,6,8]
merge(a,0,3,7)

print(a)
```
