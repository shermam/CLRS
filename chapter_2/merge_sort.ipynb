```python
import math
```

```python
# I updated the code here to consider the upper limit as non-inclusive, as it is common for the `range` function in python.
def merge(a: list[int], p: int, q: int, r: int):
  n1 = q - p
  n2 = r - q
  left = [float('inf')] * (n1+1)
  right = [float('inf')] * (n2+1)
  for i in range(n1):
    left[i]=a[p+i]
  for j in range(n2):
    right[j]=a[q+j]
  i=0
  j=0
  for k in range(p,r):
    if left[i] <= right[j]:
      a[k]=left[i]
      i+=1
    else:
      a[k]=right[j]
      j+=1

def merge_sort(a: list[int], p: int, r: int):
  if r-p>1:
    q = math.floor((p+r)/2)
    # Now that I am considering the upper limit as non-inclusive, the changes I made had another side-effect.
    # In the pseudo code in the book, arrays start at index 1 and have the upper-limit inclusive. I made some
    # modifications in the python code to have the upper limit non-inclusive. And this results in a slight difference
    # in behaviour from the book. In the pseudocode version of the algorithm, when the section of the array being
    # split has an odd number of elements, the first half would have one more element than the second. But in my
    # python version, the second half will have the extra element. Other than that, the behaviour should be similar.
    merge_sort(a, p, q)
    merge_sort(a, q, r)
    merge(a,p,q,r)
```

```python
a = [9,2,1,4,8,5,3,7,6,2]
merge_sort(a,0,len(a))
print(a)

a = [5,2,4,7,1,3,2,6]
merge_sort(a,0,len(a))
print(a)
```

```python
# [2,4,6,8,1,3,5,7] p=0 q=4 r=8
# n1 = 4; n2 = 4;
# left = [2,4,6,8]; right = [1,3,5,7]

# k=0 i=0 j=0 a=[2,4,6,8,1,3,5,7]
# k=1 i=0 j=1 a=[1,4,6,8,1,3,5,7]
# k=2 i=1 j=1 a=[1,2,6,8,1,3,5,7]
# k=3 i=1 j=2 a=[1,2,3,8,1,3,5,7]
# k=4 i=2 j=2 a=[1,2,3,4,1,3,5,7]
# k=5 i=2 j=3 a=[1,2,3,4,5,3,5,7]
# k=6 i=3 j=3 a=[1,2,3,4,5,3,6,7]
# k=7 i=3 j=3 a=[1,2,3,4,5,3,7,7]
# k=8 i=3 j=4 a=[1,2,3,4,5,3,7,7]

def merge2(a: list[int], p: int, q: int, r: int):
  n1 = q - p
  n2 = r - q
  left = [0] * (n1)
  right = [0] * (n2)
  for i in range(n1):
    left[i]=a[p+i]
  for j in range(n2):
    right[j]=a[q+j]
  i=0
  j=0
  for k in range(p,r):
    if left[i] <= right[j]:
      a[k]=left[i]
      i+=1
    else:
      a[k]=right[j]
      j+=1
    if i == n1 or j == n2:
      break

  for i1 in range(i,n1):
    a[p+j+i1] = left[i1]
  for j1 in range(j,n2):
    a[p+i+j1] = right[j1]

def merge_sort2(a: list[int], p: int, r: int):
  if r-p>1:
    q = math.floor((p+r)/2)
    merge_sort2(a, p, q)
    merge_sort2(a, q, r)
    merge2(a,p,q,r)
```

```python
a = [9,2,1,4,8,5,3,7,6,2]
merge_sort(a,0,len(a))
print(a)

a = [5,2,4,7,1,3,2,6]
merge_sort(a,0,len(a))
print(a)

a = [9,2,1,4,8,5,3,7,6,2]
merge_sort2(a,0,len(a))
print(a)

a = [2,4,6,8,1,3,5,7]
merge2(a,0,4,len(a))
print(a)

a = [5,2,4,7,1,3,2,6]
merge_sort2(a,0,len(a))
print(a)
```
