```python
import math
```

```python
# I updated the code here to consider the upper limit as non-inclusive, as it is common for the `range` function in python.
def merge(a: list[int], p: int, q: int, r: int):
  n1 = q - p
  n2 = r - q
  left = [float('inf')] * (n1+1)
  right = [float('inf')] * (n2+1)
  for i in range(n1):
    left[i]=a[p+i]
  for j in range(n2):
    right[j]=a[q+j]
  i=0
  j=0
  for k in range(p,r):
    if left[i] <= right[j]:
      a[k]=left[i]
      i+=1
    else:
      a[k]=right[j]
      j+=1

def merge_sort(a: list[int], p: int, r: int):
  if r-p>1:
    q = math.floor((p+r)/2)
    merge_sort(a, p, q)
    merge_sort(a, q, r)
    merge(a,p,q,r)
```

```python
a = [9,2,1,4,8,5,3,7,6,2]
merge_sort(a,0,len(a))
print(a)
```
