# Insertion Sort

Python indexes starting at 0 rather than 1 as in the book illustration.
This changes the range we use. 
We do from 1 inclusive to length exclusive, rather than 2 inclusive to length inclusive.
We also change the check on i in the while clause to check for >= 0 rather than just > 0.
Everything else should match the pseudocode in the book.

```python
# a = [5,2,4,6,1,3]
a = [31,41,59,26,41,58]
print(a)
```

```python
for j in range(1, len(a)):
  key = a[j]
  # Insert a[j] into the sorted sequence a[0..j-1].
  i = j-1
  while i >= 0 and a[i]>key:
    a[i+1] = a[i]
    i-=1
  a[i+1] = key
```

```python
print(a)
```
